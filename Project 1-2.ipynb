```julia
#Project 1 pkgs
using LinearAlgebra
using Plots
plotly();
```

```julia
#Problem 2 data
x=[0;1;2;3;4;5;6;7]
y=[4;3.1;2.8;2.5;2.3;2.2;2.1;2.0]
t=-1:.1:10
```

```julia
#Problem 2-a
A=[x ones(8,1)]
z=(A'A)\(A'y)
function f(x)
  return z[1]*x+z[2]
end
#calculate error vector and norm for part d
E=f.(x)-y
E_1=norm(E)
println(E_1)
#Plot graph
s=f.(t)
plot(t,s)
scatter!(x,y)
```

```julia
#Problem 2-b
A=[x.^3 x.^2 x ones(8,1)]
z=(A'A)\(A'y)
function g(x)
  return z[1]*x^3+z[2]*x^2+z[3]*x+z[4]
end
#calculate error vector and norm for part d
E=g.(x)-y
E_2=norm(E)
println(E_2)
#Plot graph
s=g.(t)
plot(t,s)
scatter!(x,y)
```

```julia
#Problem 2-c
A=[x ones(8,1)]
y=[log(4);log(3.1);log(2.8);log(2.5);log(2.3);log(2.2);log(2.1);log(2.0)]
z=(A'A)\(A'y)
z=[z[1];exp(z[2])]
function q(x)
  return z[2]*exp(z[1]*x)
end
#calculate error vector and norm for part d
y=[4;3.1;2.8;2.5;2.3;2.2;2.1;2.0]
E=q.(x)-y
E_3=norm(E)
println(E_3)
#Plot graph
s=q.(t)
plot(t,s)
scatter!(x,y)


```
