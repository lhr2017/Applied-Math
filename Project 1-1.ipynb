Functions

```julia
function bern(k,n,t)
  return binomial(n,k)*t^k*(1-t)^(n-k)
end

function bez(p)
   x=p[1:2:7]
   y=p[2:2:8]
   t=0:.01:1
   b0=bern.(0,3,t)
   b1=bern.(1,3,t)
   b2=bern.(2,3,t)
   b3=bern.(3,3,t)
   B=[b0 b1 b2 b3]
   xx=B*x
   yy=B*y
   plot!(xx,yy,legend=false)
   scatter!(x,y,legend=false)
end

function bez2(p)
  scatter(0,0,legend=false) #to clear any previous plots
  n=convert(Int,length(p)/8)
  for i=1:n
    bez(p[i,:]')
  end
  scatter!(0,0,legend=false) #to actually show the plot
end
```
